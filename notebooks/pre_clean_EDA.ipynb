# Temporarily Change dtypes for Initial EDA

```python
df['rating'].unique()
df['category'].head(4)
```

```python
df['actual_price_temp'] = df['actual_price'].replace({"₹":"",",":""}, regex=True).astype(float)
df['discount_percentage_temp'] = df['discount_percentage'].str.replace("%","").astype(float)
df['discount_price_temp'] = df['discounted_price'].replace({"₹":"",",":""}, regex=True).astype(float)
df['rating_temp']= df['rating'].replace(r'\|', np.nan, regex=True).astype(float)
df['rating_count_temp'] = df['rating_count'].str.replace(",","").astype(float)
df.info()
```

```python
temp_df = df.filter(like='temp')
temp_df.describe()
```
