```python
import numpy as np
import pandas as pd
import seaborn as sns


df = pd.read_csv(r"/resources\raw\amazon.csv")
df
```

## 1. Drop Unneeded Columns

```python
df_cleaned = df.drop(columns=['about_product','user_id','user_name','review_id','review_content','review_title','img_link','product_link'])
```

```python
df_cleaned.head()
```

## 2. Clean and Convert Object-Type Columns to Numeric-Type

```python
df_cleaned['actual_price'] = df_cleaned['actual_price'].replace({"₹":"",",":""}, regex=True).astype(float)
df_cleaned['discount_percentage'] = df_cleaned['discount_percentage'].str.replace("%","").astype(float)
df_cleaned['discounted_price'] = df_cleaned['discounted_price'].replace({"₹":"",",":""}, regex=True).astype(float)
df_cleaned['rating'] = df_cleaned['rating'].replace(r'\|', np.nan, regex=True).astype(float)
df_cleaned['rating_count'] = df_cleaned['rating_count'].str.replace(",","").astype(float)
```

```python
df_cleaned.info()
```

## 4. Correct Discounted Prices Using Discount Percentage & Actual Price

```python
df_cleaned['discounted_price'] = (df_cleaned['actual_price'] * (1-(df_cleaned['discount_percentage'] * 0.01))).round(2)
df_cleaned
```

## 5. Convert Indian Rupees to USD

```python
df_cleaned[['discounted_price','actual_price']] = (df_cleaned[['discounted_price','actual_price']] * 0.012)
df_cleaned
```

## 6. Standardize Text Data

```python
df_cleaned['product_name'] = df_cleaned['product_name'].str.strip().str.lower()
df_cleaned['category'] = df_cleaned['category'].str.strip().str.lower()
```

```python
def find_similar_names(df):
    duplicates = []
    for i, product in enumerate(df['product_name']):
        for j, other_product in enumerate(df['product_name']):
            # Skip comparing the same product with itself
            if i != j:
                # If one name is a substring of another, mark it as a potential typo
                if product in other_product or other_product in product:
                    duplicates.append((product, other_product))
    return duplicates

similar_products = find_similar_names(df_cleaned)
duplicates_df = pd.DataFrame(similar_products,columns=['product_name','similar_names'])
duplicates_df
```

```python
df_cleaned[(df_cleaned['product_name'].str.startswith('tata'))]
```

```python
df_cleaned[['main_category', 'sub_category']] = df_cleaned['category'].str.split('|', expand=True, n=1)
```

```python
df_cleaned.drop('category', axis=1, inplace=True)
```

```python
df_cleaned['sub-category'] = df_cleaned['sub-category'].str.split('|', expand=True, n=1).iloc[:,:1]
```

```python
df_cleaned
```

## 7. Handle Duplicated and Null Data

```python
df_cleaned.duplicated().sum()
```

```python
df_cleaned.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
df_cleaned = df_cleaned[df_cleaned.duplicated(keep='first') == False]
df_cleaned.duplicated().sum()
```

### A. Handle Column-specific Duplicates

```python
df_cleaned[df_cleaned.duplicated(subset=['product_id'], keep = False)]
```

```python
df_cleaned = df_cleaned.groupby('product_id').agg({
    'product_name': 'first',
    'main_category': 'first',
    'sub-category': 'first',
    'actual_price': 'mean',
    'discounted_price': 'mean',
    'discount_percentage': 'mean',
    'rating': 'mean',
    'rating_count': 'mean',
}).reset_index()
df_cleaned
```

```python
df_cleaned.product_name.duplicated().sum()
```

```python
df_cleaned = df_cleaned[~df_cleaned.duplicated(subset=['product_name'], keep = 'first')]
df_cleaned.product_name.duplicated().sum()
```
